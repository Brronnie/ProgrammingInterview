```python
import matplotlib.pyplot as plt
```

### 1. Define the derivation functions

```python
#The Derivation of S
def dS(step, subtrate, intermediate):
    a1 = 100/60
    a2 = 600/60
    lamda = a1 * 1
    fun_s = -lamda*subtrate + (a1*subtrate + a2)*intermediate
    return fun_s

#The Derivation of ES
def dES(step, subtrate, intermediate):
    b1 = 100/60
    b2 = 600/60
    b3 = 150/60
    theta = b2 + b3
    lamda = b1 * 1
    fun_es = lamda*subtrate - (theta + b1*subtrate)*intermediate
    return fun_es
```

### 2. Set the initial concentrations 

```python
h = 0.01      # time pricision
n = int(32/h)  # step number
t = []         # list of steps

E0 = 1
k3 = 150/60

temp = 0
es = [0]
s = [10]
for i in range(1, n+1):
    temp = h*i
    print('Time:', temp)
    t.append(temp)  # add initial value
```

### 3. Simulate

```python
for k in range(0, n-1):
    z0 = h * dS(t[k], s[k], es[k])
    l0 = h * dES(t[k], s[k], es[k])
    z1 = h * dS(t[k]+h/2, s[k]+z0/2, es[k]+l0/2)
    l1 = h * dES(t[k]+h/2, s[k]+z0/2, es[k]+l0/2)
    z2 = h * dS(t[k]+h/2, s[k]+z1/2, es[k]+l1/2)
    l2 = h * dES(t[k]+h/2, s[k]+z1/2, es[k]+l1/2)
    z3 = h * dS(t[k]+h, s[k]+z2, es[k]+l2)
    l3 = h * dES(t[k]+h, s[k]+z2, es[k]+l2)
    s_temp = s[k] + (z0+2*z1+2*z2+z3)/6
    es_temp = es[k] + (l0+2*l1+2*l2+l3)/6
    s.append(s_temp)
    es.append(es_temp)
```

```python
E = []
P = []
Vp = []

for item in es:
    E_value = E0 - item
    E.append(E_value)

the_sum = 0
for k in range(0, n):
    the_sum = es[k] + the_sum
    P_value = k3 * the_sum * h  # integrating discrete results
    P.append(P_value)

for item in es:
    Vp_value = k3*item
    Vp.append(Vp_value)
```

### 4. Plot the results

```python
plt.figure(dpi=200)
plt.axis([0, 35, 0, 15])
plt.xlabel('Time(s)')
plt.ylabel('Concentration(uM)')
l_1, = plt.plot(t, E, color='blue')
l_2, = plt.plot(t, s, color='red')
l_3, = plt.plot(t, es, color='purple')
l_4, = plt.plot(t, P, color='green')
plt.legend(loc='best', handles=[l_1, l_2, l_3, l_4], fontsize=6,
           labels=['E', 'S', 'ES', 'P'])
plt.show()

plt.figure(dpi=200)
plt.axis([0, 15, 0, 2])
plt.xlabel('Concentration of S(uM)')
plt.ylabel('Change rate of P(uM/s)')
l_5, = plt.plot(s, Vp, color='red')
plt.show()
```
